```python
import pandas as pd
import numpy as np
```

```python
performance = ["G1", "G2", "G3"]
student_performance = data[performance]
```

```python
student_performance
```

```python
student_performance = student_performance / student_performance.max() #normalizing performance (0-1)
```

```python
student_performance
```

```python
student_performance.isnull().sum()
```

```python
student_subject_matrix = student_performance.T #student-subject matrix
```

```python
student_subject_matrix
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity_matrix = cosine_similarity(student_performance)
```

```python
def recommend_topics(student_id, n_recommendations=2):
    low_scores = student_performance.iloc[student_id][student_performance.iloc[student_id] < 0.5].index
    similarities = similarity_matrix[student_id]
    recommended_scores = {}
    for similar_student_id, similarity in enumerate(similarities):
        if similar_student_id != student_id:
            for topic in low_scores:
                if student_performance.iloc[similar_student_id][topic] > 0.7:
                    recommended_scores[topic] = recommended_scores.get(topic, 0) + similarity
    recommended_topics = sorted(recommended_scores, key=recommended_scores.get, reverse=True)
    return recommended_topics[:n_recommendations]
```

```python
student_id = 9
recommendations = recommend_topics(student_id, n_recommendations=2)
print(f"Recommended topics for Student {student_id}: {recommendations}")
```
